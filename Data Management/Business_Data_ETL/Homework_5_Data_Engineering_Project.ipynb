## Making 10 cleaning or transforming operations

Operation 1: Remove column 'nickname'

```python
# Remove column nickname
ufc_fighters_statistics_pd.drop('nickname', axis=1, inplace=True)
print(ufc_fighters_statistics_pd.head())
```

Operation 2: Creating a new column 'win%' by calculating the win percentage

```python
# Calculate win percentage and create a new column
ufc_fighters_statistics_pd['win%'] = (ufc_fighters_statistics_pd['wins'] / (ufc_fighters_statistics_pd['wins'] + ufc_fighters_statistics_pd['losses'])) * 100

# Print the updated DataFrame with the new 'win%' column
print(ufc_fighters_statistics_pd.head())
```

Operation 3: Renaming column the 'win%' column to 'win_percentage'

```python
# Rename the 'win%' column to 'win_percentage'
ufc_fighters_statistics_pd.rename(columns={'win%': 'win_percentage'}, inplace=True)

# Print the DataFrame with the renamed column
print(ufc_fighters_statistics_pd.head())
```

Operation 4: Capitalizing the first letter of each column name

```python
# Capitalize column names
ufc_fighters_statistics_pd.columns = ufc_fighters_statistics_pd.columns.str.title()

# Print the DataFrame with updated column names
print(ufc_fighters_statistics_pd.head())
```

Operation 5: Replacing the missing value with another value of column 'Stance'

```python
# Replace missing values in 'Stance' column with 'unannounced'
ufc_fighters_statistics_pd['Stance'] = ufc_fighters_statistics_pd['Stance'].fillna('unannounced')

# Print the DataFrame to verify the changes
print(ufc_fighters_statistics_pd.head())
```

Operation 6: Grouping data based on 'Southpaw' category of 'Stance' column

```python
southpaw_group = ufc_fighters_statistics_pd.groupby('Stance').get_group('Southpaw')

# Print the grouped data for 'Southpaw'
print(southpaw_group)
```

Operation 7: Dropping rows with missing values

```python
# Drop rows with any missing values
ufc_fighters_statistics_pd.dropna(inplace=True)

# Print the DataFrame after dropping null values
print(ufc_fighters_statistics_pd.head())
```

Operation 8: Spliting 'Date_Of_Birth' column into separate 'Birth', 'Year', and 'Day' columns

```python
# Split 'Date_Of_Birth' column into 'Birth', 'Year', and 'Day'
ufc_fighters_statistics_pd[['Birth', 'Year', 'Day']] = ufc_fighters_statistics_pd['Date_Of_Birth'].str.split('-', expand=True)

print(ufc_fighters_statistics_pd.head())
```

Operation 9: Categorization of generation based on birth year

```python
# Create a function to categorize birth years
def categorize_generation(year):
    if 1980 <= year < 1990:
        return "80's"
    elif 1990 <= year < 2000:
        return "90's"
    elif 2000 <= year < 2010:
        return "GenZ"
    else:
        return "Old"

# Apply the function to create a new column 'Generation'
ufc_fighters_statistics_pd['Generation'] = ufc_fighters_statistics_pd['Year'].astype(int).apply(categorize_generation)

# Print the DataFrame with the new 'Generation' column
print(ufc_fighters_statistics_pd)
```

Operation 10: Filtering the players who have fewer than 5 looses

```python
# Filtering the players who have fewer than 5 looses
filtered_data = ufc_fighters_statistics_pd[ufc_fighters_statistics_pd['Losses'] < 5]
print(filtered_data)
```

```python
# Url of the page to webscrape
oer_url = 'https://guides.lib.uh.edu/OER/ATIP/awardees'
# google_cache = 'http://webcache.googleusercontent.com/search?q=cache:'
```

```python
import requests
from bs4 import BeautifulSoup

response = requests.get(oer_url)
soup = BeautifulSoup(response.text, "html.parser")
```

```python
response.text
```

```python
years = soup.select("h2.s-lib-box-title")
years
```

```python
# Extracting years

year_texts = [year.get_text(strip=True) for year in years]
print(year_texts)
```

```python
years = ['ATIP Awardees', '2022', '2020', '2019', '2018', 'Contact']

# Filter out unwanted elements
filtered_years = [year for year in years if year.isdigit()]

# Display the filtered list of years
print(filtered_years)

```

```python
tables = soup.select("table.striped")
tables
```

```python
import pandas as pd

# List of years in the correct order
filtered_years = ['2022', '2020', '2019', '2018']

```

```python
# Function to extract data from a table row
def extract_data(row, year):
    cols = row.find_all('td')
    return {
        'Name': cols[0].get_text(strip=True),
        'College': cols[1].get_text(strip=True),
        'Course': cols[2].get_text(strip=True),
        'Year': year
    }
```

```python
# Extract data from each table and concatenate into a list
data_frames = []
for idx, table in enumerate(tables):
    rows = table.find_all('tr')
    temp_data = [extract_data(row, filtered_years[idx]) for row in rows[1:]]
    data_frames.append(pd.DataFrame(temp_data))

```

```python
# Concatenate all DataFrames into one
oer_table = pd.concat(data_frames, ignore_index=True)
print(oer_table.head())
print(oer_table.info())
```
